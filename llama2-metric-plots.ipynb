# Accuracy, precission and recall

```python
def plot_barchart(labels,list,list_ft,task):

    list_rounded = [round(num, 2) for num in list]
    list_ft_rounded = [round(num, 2) for num in list_ft]

    fig = go.Figure(data=[
        go.Bar(name='Fine-tuned model', x=labels, y=list_ft, marker_color='blue',
               text=list_ft_rounded, textposition='inside', textfont=dict(color='white')),
        go.Bar(name='Base model', x=labels, y=list, marker_color='orange',
               text=list_rounded, textposition='inside', textfont=dict(color='white'))
    ])

    # Update the layout
    fig.update_layout(  
        xaxis_title='Categories',
        yaxis_title=task,
        barmode='group',
        xaxis_tickangle=-45,
        xaxis_tickfont=dict(size=22),
        yaxis_tickfont=dict(size=22),
        yaxis_title_font=dict(size=22),
        xaxis_title_font=dict(size=22),
        legend=dict(
            orientation="h",  # Horizontal legend
            yanchor="bottom",  # Align to the bottom of the legend
            y=1.1,  # Place it above the plot
            xanchor="center",  # Center the legend
            x=0.5,  # Place it in the middle of the plot
            font=dict(size=22)  
        )
    )

    # Show the figure
    return fig.show()

plot_barchart(labels,accuracy,accuracy_ft,'Accuracy')
plot_barchart(labels,precision,precision_ft,'Precission')
plot_barchart(labels,recall,recall_ft,'Recall')
```
